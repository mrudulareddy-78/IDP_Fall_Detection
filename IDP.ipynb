```python
# ============================================================
# CELL 1 — Imports
# ============================================================
import os, glob, re, warnings, time, gc
import numpy as np
import pandas as pd
from scipy.signal import butter, filtfilt
from sklearn.model_selection import train_test_split
from sklearn.metrics import (classification_report, confusion_matrix,
                              f1_score, precision_score, recall_score,
                              accuracy_score)
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import seaborn as sns

warnings.filterwarnings("ignore")
np.random.seed(42)
tf.random.set_seed(42)

print("TensorFlow:", tf.__version__)
print("GPU:", tf.config.list_physical_devices('GPU'))
print("✅ Imports OK")


# ============================================================
# CELL 2 — Upload & Extract ZIPs
# ============================================================
import zipfile
from google.colab import files

DATASET_ROOT = "/content/datasets"
os.makedirs(DATASET_ROOT, exist_ok=True)

print("Upload archive.zip (SisFall) AND archive(1).zip (KFall)")
uploaded = files.upload()

for fname in uploaded.keys():
    try:
        with zipfile.ZipFile(fname, 'r') as zf:
            zf.extractall(DATASET_ROOT)
        print("✓ Extracted:", fname)
    except Exception as e:
        print("⚠ Error:", fname, e)

print("Done. Dataset at:", DATASET_ROOT)


# ============================================================
# CELL 3 — Auto-detect Paths
# ============================================================
def find_dir_with_files(root, ext):
    for dp, _, fnames in os.walk(root):
        if any(f.endswith(ext) for f in fnames):
            return dp
    return None

SISFALL_ROOT = find_dir_with_files(DATASET_ROOT, ".txt")
KFALL_ROOT   = find_dir_with_files(DATASET_ROOT, ".csv")

print("SisFall root:", SISFALL_ROOT)
print("KFall root  :", KFALL_ROOT)


# ============================================================
# CELL 4 — Global Parameters
# ============================================================
SISFALL_FS     = 200
KFALL_FS       = 100
TARGET_FS      = 100
BUTTER_ORDER   = 6
CUTOFF_HZ      = 20.0
WINDOW         = 300        # 3s × 100Hz
STEP           = 150        # 50% overlap
KFALL_FALL_IDS = set(range(20, 35))

# ── Memory control ────────────────────────────────────────────
# Limit windows PER TRIAL to cap RAM usage.
# Each window = 300×8×4 bytes = ~9.6 KB
# 5000 windows × 9.6 KB ≈ 48 MB per trial → fine
MAX_WINDOWS_PER_TRIAL = 20   # max windows kept per trial (set None to keep all)

print(f"Window : {WINDOW} samples = {WINDOW/TARGET_FS:.1f}s")
print(f"Step   : {STEP} samples")


# ============================================================
# CELL 5 — Pre-processing Functions
# ============================================================
def resample_signal(sig, orig_fs):
    if orig_fs == TARGET_FS:
        return sig
    n_orig   = len(sig)
    n_target = int(n_orig * TARGET_FS / orig_fs)
    x_orig   = np.linspace(0, 1, n_orig)
    x_new    = np.linspace(0, 1, n_target)
    out = np.zeros((n_target, sig.shape[1]), dtype=np.float32)
    for ch in range(sig.shape[1]):
        out[:, ch] = np.interp(x_new, x_orig, sig[:, ch])
    return out

def butter_lowpass(sig):
    b, a = butter(BUTTER_ORDER, CUTOFF_HZ / (0.5 * TARGET_FS),
                  btype='low', analog=False)
    return filtfilt(b, a, sig, axis=0).astype(np.float32)

def add_magnitude_features(sig):
    acc_mag  = np.linalg.norm(sig[:, :3], axis=1, keepdims=True)
    gyro_mag = np.linalg.norm(sig[:, 3:6], axis=1, keepdims=True)
    return np.hstack([sig, acc_mag, gyro_mag]).astype(np.float32)

def sliding_window(sig):
    wins = []
    for i in range(0, len(sig) - WINDOW + 1, STEP):
        wins.append(sig[i : i + WINDOW])
    # ── Limit windows per trial to control RAM ────────────────
    if MAX_WINDOWS_PER_TRIAL and len(wins) > MAX_WINDOWS_PER_TRIAL:
        idx = np.linspace(0, len(wins)-1, MAX_WINDOWS_PER_TRIAL, dtype=int)
        wins = [wins[i] for i in idx]
    return wins

print("✅ Pre-processing functions defined.")


# ============================================================
# CELL 6 — Loaders
# ============================================================
def load_sisfall(root):
    records = []
    for fp in glob.glob(os.path.join(root, "**/*.txt"), recursive=True):
        try:
            label = 1 if os.path.basename(fp).upper().startswith('F') else 0
            rows  = []
            with open(fp) as f:
                for line in f:
                    v = line.strip().replace(';', '').split(',')
                    if len(v) >= 6:
                        rows.append([float(x) for x in v[:6]])
            if len(rows) > 10:
                records.append((np.array(rows, dtype=np.float32)/100.0,
                                label, SISFALL_FS))
        except:
            pass
    print(f"  SisFall: {len(records)} trials")
    return records

def load_kfall(root):
    records = []
    for fp in glob.glob(os.path.join(root, "**/*.csv"), recursive=True):
        try:
            df = pd.read_csv(fp)
            col_sets = [
                ["AccX","AccY","AccZ","GyrX","GyrY","GyrZ"],
                ["acc_x","acc_y","acc_z","gyr_x","gyr_y","gyr_z"],
                ["Acc_X","Acc_Y","Acc_Z","Gyr_X","Gyr_Y","Gyr_Z"],
            ]
            sig = None
            for cols in col_sets:
                if all(c in df.columns for c in cols):
                    sig = df[cols].values.astype(np.float32)
                    break
            if sig is None:
                num_cols = df.select_dtypes(include=np.number).columns[:6]
                if len(num_cols) < 6:
                    continue
                sig = df[num_cols].values.astype(np.float32)
            if len(sig) < 10:
                continue
            m = re.findall(r'[Tt](\d+)', os.path.basename(fp))
            label = 1 if (m and int(m[0]) in KFALL_FALL_IDS) else 0
            records.append((sig, label, KFALL_FS))
        except:
            pass
    print(f"  KFall: {len(records)} trials")
    return records


# ============================================================
# CELL 7 — Load Datasets
# ============================================================
print("Loading SisFall...")
sisfall_recs = load_sisfall(SISFALL_ROOT) if SISFALL_ROOT else []

print("Loading KFall...")
kfall_recs   = load_kfall(KFALL_ROOT) if KFALL_ROOT else []

all_records = sisfall_recs + kfall_recs
all_labels  = [r[1] for r in all_records]

print(f"\nTotal : {len(all_records)} trials | Falls: {sum(all_labels)}")
if not all_records:
    raise Exception("❌ No records loaded")


# ============================================================
# CELL 8 — Trial-level Split BEFORE Windowing
# ============================================================
train_recs, test_recs = train_test_split(
    all_records, test_size=0.20,
    stratify=all_labels, random_state=42
)
train_labels_only = [r[1] for r in train_recs]
train_recs, val_recs = train_test_split(
    train_recs, test_size=0.15,
    stratify=train_labels_only, random_state=42
)

# Free raw list — no longer needed
del all_records, sisfall_recs, kfall_recs
gc.collect()

print(f"Train: {len(train_recs)} | Val: {len(val_recs)} | Test: {len(test_recs)}")


# ============================================================
# CELL 9 — Memory-Safe Process Function
# ============================================================
# ✅ Processes one split at a time and immediately frees raw signals.
# ✅ Uses np.float16 storage to halve RAM (converted to float32 for training).

def process_records(recs, desc=""):
    """
    Process trials → windows.
    Stores as float16 to save RAM; convert to float32 when feeding model.
    """
    X_list, y_list = [], []
    for i, (sig, label, fs) in enumerate(recs):
        try:
            sig = resample_signal(sig, fs)
            if len(sig) > BUTTER_ORDER * 3:
                sig = butter_lowpass(sig)
            sig = add_magnitude_features(sig)
            for w in sliding_window(sig):
                X_list.append(w.astype(np.float16))  # float16 → half the RAM
                y_list.append(label)
            del sig  # free immediately
        except:
            pass
        # Progress every 200 trials
        if (i+1) % 200 == 0:
            print(f"  {desc}: {i+1}/{len(recs)} trials processed...")

    X = np.array(X_list, dtype=np.float16)
    y = np.array(y_list,  dtype=np.int32)
    del X_list, y_list
    gc.collect()
    print(f"  {desc}: {X.shape} | {X.nbytes/1024/1024:.1f} MB")
    return X, y


print("Processing train...")
X_train_raw, y_train = process_records(train_recs, "Train")

print("Processing val...")
X_val_raw, y_val     = process_records(val_recs,   "Val")

print("Processing test...")
X_test_raw, y_test   = process_records(test_recs,  "Test")

# Free trial lists
del train_recs, val_recs, test_recs
gc.collect()

print(f"\nX_train: {X_train_raw.shape}  Falls: {np.sum(y_train==1)}")
print(f"X_val  : {X_val_raw.shape}    Falls: {np.sum(y_val==1)}")
print(f"X_test : {X_test_raw.shape}   Falls: {np.sum(y_test==1)}")


# ============================================================
# CELL 10 — Z-score Normalization (train stats only)
# ============================================================
# Compute on float32 view for precision, store result as float16

train_f32  = X_train_raw.astype(np.float32)
train_mean = train_f32.mean(axis=(0, 1), keepdims=True)
train_std  = train_f32.std(axis=(0, 1),  keepdims=True)
train_std  = np.where(train_std < 1e-6, 1.0, train_std)
del train_f32
gc.collect()

np.save('/content/norm_mean.npy', train_mean)
np.save('/content/norm_std.npy',  train_std)

X_train = ((X_train_raw.astype(np.float32) - train_mean) / train_std).astype(np.float16)
del X_train_raw; gc.collect()

X_val   = ((X_val_raw.astype(np.float32)   - train_mean) / train_std).astype(np.float16)
del X_val_raw;   gc.collect()

X_test  = ((X_test_raw.astype(np.float32)  - train_mean) / train_std).astype(np.float16)
del X_test_raw;  gc.collect()

print("✅ Normalized.")
print(f"RAM: train={X_train.nbytes/1024/1024:.0f}MB  "
      f"val={X_val.nbytes/1024/1024:.0f}MB  "
      f"test={X_test.nbytes/1024/1024:.0f}MB")


# ============================================================
# CELL 11 — Balance Training Data
# ============================================================
fall_idx  = np.where(y_train == 1)[0]
nfall_idx = np.where(y_train == 0)[0]
n_maj     = max(len(fall_idx), len(nfall_idx))

print(f"Before → Falls:{len(fall_idx)}  Non-falls:{len(nfall_idx)}")

if len(fall_idx) < len(nfall_idx):
    fall_bal  = np.random.choice(fall_idx,  n_maj, replace=True)
    final_idx = np.concatenate([fall_bal, nfall_idx])
else:
    nfall_bal = np.random.choice(nfall_idx, n_maj, replace=True)
    final_idx = np.concatenate([fall_idx, nfall_bal])

np.random.shuffle(final_idx)
X_train = X_train[final_idx]
y_train = y_train[final_idx]
gc.collect()

print(f"After  → Falls:{np.sum(y_train==1)}  Non-falls:{np.sum(y_train==0)}")
print(f"X_train: {X_train.shape}  {X_train.nbytes/1024/1024:.0f} MB")


# ============================================================
# CELL 12 — Gaussian Noise Augmentation (train only)
# ============================================================
noise = np.random.normal(0, 0.03, X_train.shape).astype(np.float16)
X_train = X_train + noise
del noise
gc.collect()
print("✅ Noise augmentation done.")


# ============================================================
# CELL 13 — tf.data Pipeline (avoids loading all float32 at once)
# ============================================================
# Convert float16 → float32 lazily inside tf.data so GPU gets
# float32 but CPU RAM stays at float16.

BATCH_SIZE = 64

def make_dataset(X, y, shuffle=False):
    """
    Wraps float16 numpy arrays in tf.data.
    Cast to float32 happens on the GPU-side graph, not in RAM.
    """
    ds = tf.data.Dataset.from_tensor_slices(
        (X.astype(np.float32), y.astype(np.float32))
    )
    if shuffle:
        ds = ds.shuffle(buffer_size=4096, seed=42)
    ds = ds.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
    return ds

train_ds = make_dataset(X_train, y_train, shuffle=True)
val_ds   = make_dataset(X_val,   y_val)
test_ds  = make_dataset(X_test,  y_test)

print("✅ tf.data pipelines ready.")


# ============================================================
# CELL 14 — Model: Conv1D + LSTM
# ============================================================
N_CHANNELS = X_train.shape[2]   # 8

def build_model(input_shape=(WINDOW, N_CHANNELS)):
    inp = layers.Input(shape=input_shape, name='imu_input')

    # CNN Block 1
    x = layers.Conv1D(64, 5, padding='same', activation='relu')(inp)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling1D(2)(x)
    x = layers.Dropout(0.2)(x)

    # CNN Block 2
    x = layers.Conv1D(128, 3, padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling1D(2)(x)
    x = layers.Dropout(0.2)(x)

    # CNN Block 3
    x = layers.Conv1D(64, 3, padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.2)(x)

    # LSTM Block
    x = layers.LSTM(128, return_sequences=True,
                    activation='tanh',
                    recurrent_activation='sigmoid')(x)
    x = layers.Dropout(0.3)(x)
    x = layers.LSTM(64, return_sequences=False,
                    activation='tanh',
                    recurrent_activation='sigmoid')(x)
    x = layers.Dropout(0.3)(x)

    # Classifier
    x   = layers.Dense(64, activation='relu')(x)
    x   = layers.Dropout(0.2)(x)
    out = layers.Dense(1, activation='sigmoid', name='fall_prob')(x)

    m = models.Model(inputs=inp, outputs=out, name='CNN_LSTM_FallDetector')
    return m

model = build_model()
model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-3),
    loss='binary_crossentropy',
    metrics=['accuracy',
             tf.keras.metrics.Precision(name='precision'),
             tf.keras.metrics.Recall(name='recall'),
             tf.keras.metrics.AUC(name='auc')]
)
model.summary()


# ============================================================
# CELL 15 — Train
# ============================================================
callbacks = [
    tf.keras.callbacks.EarlyStopping(
        monitor='val_loss', patience=10,
        restore_best_weights=True, verbose=1),
    tf.keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss', factor=0.5,
        patience=5, min_lr=1e-6, verbose=1),
    tf.keras.callbacks.ModelCheckpoint(
        '/content/best_fall_model.keras',
        monitor='val_loss', save_best_only=True, verbose=0)
]

t0 = time.time()
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=60,
    callbacks=callbacks,
    verbose=1
)
print(f"\nTraining time: {(time.time()-t0)/60:.1f} min")


# ============================================================
# CELL 16 — Training Curves
# ============================================================
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].plot(history.history['loss'],         label='Train Loss')
axes[0].plot(history.history['val_loss'],     label='Val Loss')
axes[0].set_title('Loss'); axes[0].legend()
axes[1].plot(history.history['accuracy'],     label='Train Acc')
axes[1].plot(history.history['val_accuracy'], label='Val Acc')
axes[1].set_title('Accuracy'); axes[1].legend()
plt.tight_layout()
plt.savefig('/content/training_curves.png', dpi=150)
plt.show()


# ============================================================
# CELL 17 — Best Threshold from VAL Set
# ============================================================
model = tf.keras.models.load_model('/content/best_fall_model.keras')

# Predict in batches to avoid OOM on val set
y_val_prob = model.predict(val_ds, verbose=0).flatten()

thresholds = np.arange(0.30, 0.71, 0.01)
f1_scores  = [
    f1_score(y_val, (y_val_prob >= t).astype(int), zero_division=0)
    for t in thresholds
]
best_thresh = float(thresholds[np.argmax(f1_scores)])
print(f"Best threshold: {best_thresh:.2f}  (val F1={max(f1_scores):.4f})")

plt.figure(figsize=(8, 4))
plt.plot(thresholds, f1_scores)
plt.axvline(best_thresh, color='r', linestyle='--',
            label=f'Best={best_thresh:.2f}')
plt.xlabel('Threshold'); plt.ylabel('Val F1')
plt.title('Threshold vs F1 — Validation Set')
plt.legend(); plt.tight_layout()
plt.savefig('/content/threshold_curve.png', dpi=150)
plt.show()


# ============================================================
# CELL 18 — Evaluate on Test Set
# ============================================================
y_test_prob = model.predict(test_ds, verbose=0).flatten()
y_pred      = (y_test_prob >= best_thresh).astype(int)

print("\n📊 Classification Report")
print("=" * 55)
print(classification_report(y_test, y_pred,
                             target_names=['Non-Fall','Fall']))

cm = confusion_matrix(y_test, y_pred)
fig, ax = plt.subplots(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Non-Fall','Fall'],
            yticklabels=['Non-Fall','Fall'], ax=ax)
ax.set_xlabel('Predicted'); ax.set_ylabel('Actual')
ax.set_title('Confusion Matrix — Test Set')
plt.tight_layout()
plt.savefig('/content/confusion_matrix.png', dpi=150)
plt.show()

acc  = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred, zero_division=0)
rec  = recall_score(y_test, y_pred, zero_division=0)
f1   = f1_score(y_test, y_pred, zero_division=0)

print(f"Accuracy  : {acc*100:.2f}%")
print(f"Precision : {prec*100:.2f}%")
print(f"Recall    : {rec*100:.2f}%  ← missed falls: {(1-rec)*100:.2f}%")
print(f"F1 Score  : {f1*100:.2f}%")

t0 = time.time()
_ = model.predict(X_test[:50].astype(np.float32), batch_size=50)
latency_ms = (time.time()-t0)/50*1000
print(f"Avg latency: {latency_ms:.1f} ms/window")


# ============================================================
# CELL 19 — Consistency Check (Firmware Simulation)
# ============================================================
def consistency_check(probs, threshold, n_consec=3):
    preds = (probs >= threshold).astype(int)
    alerts, streak = np.zeros_like(preds), 0
    for i, p in enumerate(preds):
        streak = streak + 1 if p == 1 else 0
        if streak >= n_consec:
            alerts[i] = 1
    return alerts

alerts = consistency_check(y_test_prob, best_thresh, n_consec=3)
print(f"Raw fall predictions  : {np.sum(y_pred)}")
print(f"Alerts after 3-consec : {np.sum(alerts)}")
print(f"True falls in test    : {np.sum(y_test)}")


# ============================================================
# CELL 20 — Save Keras Model
# ============================================================
model.save('/content/fall_model.keras')
print("✅ Keras model saved.")


# ============================================================
# CELL 21 — Float32 TFLite Export
# ============================================================
converter   = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_fp32 = converter.convert()
with open('/content/fall_model_fp32.tflite','wb') as f:
    f.write(tflite_fp32)
fp32_kb = len(tflite_fp32)/1024
print(f"Float32 TFLite: {fp32_kb:.1f} KB")


# ============================================================
# CELL 22 — Int8 Post-Training Quantization
# ============================================================
# Use a small calibration sample — NOT all of X_train
CALIB_N = min(200, len(X_train))
calib_X = X_train[:CALIB_N].astype(np.float32)

def rep_data():
    for i in range(CALIB_N):
        yield [calib_X[i:i+1]]

converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations             = [tf.lite.Optimize.DEFAULT]
converter.representative_dataset    = rep_data
converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS_INT8]
converter.inference_input_type      = tf.int8
converter.inference_output_type     = tf.int8

tflite_int8 = converter.convert()
with open('/content/fall_model_int8.tflite','wb') as f:
    f.write(tflite_int8)

del calib_X; gc.collect()

int8_kb = len(tflite_int8)/1024
print(f"Float32 : {fp32_kb:.1f} KB")
print(f"Int8    : {int8_kb:.1f} KB  ({(1-int8_kb/fp32_kb)*100:.0f}% smaller)")


# ============================================================
# CELL 23 — Verify Int8 Accuracy (small sample)
# ============================================================
interp = tf.lite.Interpreter(model_path='/content/fall_model_int8.tflite')
interp.allocate_tensors()

inp_det = interp.get_input_details()[0]
out_det = interp.get_output_details()[0]
inp_scale, inp_zp = inp_det['quantization']
out_scale, out_zp = out_det['quantization']

print(f"Input  : {inp_det['dtype']} scale={inp_scale:.6f} zp={inp_zp}")
print(f"Output : {out_det['dtype']} scale={out_scale:.6f} zp={out_zp}")

n_verify   = min(200, len(X_test))
y_tfl_pred = []

for i in range(n_verify):
    x_fp32 = X_test[i:i+1].astype(np.float32)
    x_int8 = np.round(x_fp32 / inp_scale + inp_zp).clip(-128,127).astype(np.int8)
    interp.set_tensor(inp_det['index'], x_int8)
    interp.invoke()
    raw  = interp.get_tensor(out_det['index'])[0][0]
    prob = (float(raw) - out_zp) * out_scale
    y_tfl_pred.append(1 if prob >= best_thresh else 0)

y_tfl_pred = np.array(y_tfl_pred)
print(f"\nInt8 TFLite on {n_verify} samples:")
print(classification_report(y_test[:n_verify], y_tfl_pred,
                             target_names=['Non-Fall','Fall']))


# ============================================================
# CELL 24 — Generate model.h
# ============================================================
import subprocess
result = subprocess.run(
    ['xxd', '-i', '/content/fall_model_int8.tflite'],
    capture_output=True, text=True
)

header = f"""// ================================================================
// Fall Detection Model — Auto-generated C++ Header
// ECE Firmware Team — TFLite for Microcontrollers
// ================================================================
// Preprocessing (replicate EXACTLY in C++):
//   1. Resample IMU to {TARGET_FS} Hz
//   2. Butterworth LP filter: order={BUTTER_ORDER}, cutoff={CUTOFF_HZ}Hz, zero-phase
//   3. ch7 = norm(acc_xyz),  ch8 = norm(gyro_xyz)
//   4. Z-score per channel using norm_mean/norm_std arrays
//
// Input  : int8[1][{WINDOW}][{N_CHANNELS}]
// Output : int8[1][1]
// Input  quant : scale={inp_scale:.8f}, zp={inp_zp}
// Output quant : scale={out_scale:.8f}, zp={out_zp}
// Threshold    : {best_thresh:.2f}
//
// C++ dequantize:
//   float prob = (out_int8 - ({out_zp})) * {out_scale:.8f}f;
//   int   fall = (prob >= {best_thresh:.2f}f) ? 1 : 0;
// ================================================================

#ifndef FALL_MODEL_H
#define FALL_MODEL_H

{result.stdout}

#endif  // FALL_MODEL_H
"""

with open('/content/model.h', 'w') as f:
    f.write(header)
print("✅ model.h generated")


# ============================================================
# CELL 25 — Model Datasheet
# ============================================================
datasheet = f"""
╔══════════════════════════════════════════════════════════════╗
║        MODEL DATASHEET — Fall Detection System               ║
║        Handover Document for ECE Firmware Team               ║
╚══════════════════════════════════════════════════════════════╝

─── DATASETS ──────────────────────────────────────────────────
  SisFall : 19 fall + 15 ADL types   | original 200 Hz
  KFall   : Task 20–34=Fall, 1–19=ADL | original 100 Hz

─── PREPROCESSING — REPLICATE IN C++ ──────────────────────────
  1. Resample    → {TARGET_FS} Hz (linear interp)
  2. Butterworth → order={BUTTER_ORDER}, cutoff={CUTOFF_HZ}Hz, zero-phase (fwd+bwd)
  3. Features    → ch7=acc_mag, ch8=gyro_mag  (total 8 ch)
  4. Z-score     → (x - mean) / std  per channel
       mean file : norm_mean.npy  shape=(1,1,8)
       std  file : norm_std.npy   shape=(1,1,8)

─── HARDWARE ──────────────────────────────────────────────────
  IMU rate      : {TARGET_FS} Hz  ← set in hardware register
  Channels      : AccX AccY AccZ GyrX GyrY GyrZ

─── WINDOWING ─────────────────────────────────────────────────
  Window        : {WINDOW} samples = {WINDOW/TARGET_FS:.1f} s
  Step          : {STEP} samples = {STEP/TARGET_FS:.1f} s
  New prediction every {STEP/TARGET_FS*1000:.0f} ms

─── TENSORS ───────────────────────────────────────────────────
  Input  shape  : [1, {WINDOW}, {N_CHANNELS}]  dtype: int8
  Output shape  : [1, 1]        dtype: int8
  Input  quant  : scale={inp_scale:.8f}  zp={inp_zp}
  Output quant  : scale={out_scale:.8f}  zp={out_zp}
  Threshold     : {best_thresh:.2f}
  Labels        : 0=Non-Fall  1=Fall

─── PERFORMANCE ───────────────────────────────────────────────
  Accuracy  : {acc*100:.2f}%
  Precision : {prec*100:.2f}%
  Recall    : {rec*100:.2f}%  (missed falls: {(1-rec)*100:.2f}%)
  F1 Score  : {f1*100:.2f}%
  Latency   : ~{latency_ms:.1f} ms/window (GPU; measure on MCU)

─── MODEL SIZE ────────────────────────────────────────────────
  Float32   : {fp32_kb:.1f} KB
  Int8      : {int8_kb:.1f} KB  ({(1-int8_kb/fp32_kb)*100:.0f}% smaller)

─── FIRMWARE RULES ────────────────────────────────────────────
  Consistency   : 3 consecutive fall preds → alert
  Cancel timer  : 10–20 s countdown; motion/button cancels
  Duty cycle    : infer every {STEP/TARGET_FS*1000:.0f} ms

─── DELIVERABLES ──────────────────────────────────────────────
  model.h                  ← include in firmware
  fall_model_int8.tflite   ← binary
  norm_mean.npy            ← 8 mean values
  norm_std.npy             ← 8 std values
"""

print(datasheet)
with open('/content/MODEL_DATASHEET.txt','w') as f:
    f.write(datasheet)
print("✅ Datasheet saved.")


# ============================================================
# CELL 26 — Download All Deliverables
# ============================================================
from google.colab import files

deliverables = [
    '/content/fall_model_int8.tflite',
    '/content/fall_model_fp32.tflite',
    '/content/fall_model.keras',
    '/content/model.h',
    '/content/norm_mean.npy',
    '/content/norm_std.npy',
    '/content/MODEL_DATASHEET.txt',
    '/content/confusion_matrix.png',
    '/content/training_curves.png',
    '/content/threshold_curve.png',
]

for fp in deliverables:
    if os.path.exists(fp):
        files.download(fp)
        print(f"  ✓ {os.path.basename(fp)}")
    else:
        print(f"  ⚠ Missing: {fp}")
```
